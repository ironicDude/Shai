#About Dataset
salaries dataset generally provides information about the employees of an organization in relation to their compensation. It typically includes details such as how much each employee is paid (their salary), their job titles, the departments they work in, and possibly additional information like their level of experience, education, and employment history within the organization.

# About Features

1. **Employee ID**: A unique identifier for each employee. Usually a numeric or alphanumeric code.

2. **Name**: The full name of the employee. This might be split into first and last names in some datasets.

3. **Department**: The department or division within the organization where the employee works.

4. **Job Title**: The official title or position held by the employee in the organization.

5. **Salary**: The amount of money the employee earns. This could be represented as an annual salary, hourly wage, or monthly pay.

6. **Hire Date**: The date when the employee started working at the organization. Useful for calculating tenure or experience.

7. **Gender**: The gender of the employee. This might be included in some datasets for diversity and equality analysis.

8. **Years of Experience**: The total number of years the employee has worked in their field or profession.

9. **Education Level**: The highest level of education attained by the employee, such as a high school diploma, bachelor's degree, etc.

10. **Performance Rating**: If available, this could be a measure of the employee's performance in their role, often based on annual review scores.

11. **Location**: The geographical location or office where the employee works, if the organization has multiple locations.

```python
import pandas as pd
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt
```

```python
# Load your dataset
df = pd.read_csv('/content/content/MyDrive/SalariesX.csv')
```

# First request

```python
print('number of columns: ', len(df.columns))
print('number of rows: ', len(df))
print('data types: ', df.dtypes)
df.isnull()
```

# Second Request

```python
print('mean salary: ', df.TotalPayBenefits.mean())
print('median salary: ', df.TotalPayBenefits.median())
print('mode salary: ', df.TotalPayBenefits.mode())
print('minimum salary: ', df.TotalPayBenefits.min())
print('maximum salary: ', df.TotalPayBenefits.max())
print('Range of salaries: ', df.TotalPayBenefits.max() - df.TotalPayBenefits.min())
print('Standard deviation of salaries: ', df.TotalPayBenefits.std())
```

# Third Request

```python
df.head()
```

Replace the NaN values in the columns Benefits, BasePay, OvertimePay, OtherPay, TotalPay, and TotalPayBenefits with 0 because NaN indicates zero payment in this context.

```python
df['Benefits'] = df['Benefits'].fillna(0)
df['BasePay'] = df['BasePay'].fillna(0)
df['OvertimePay'] = df['OvertimePay'].fillna(0)
df['OtherPay'] = df['OtherPay'].fillna(0)
df['TotalPay'] = df['TotalPay'].fillna(0)
df['TotalPayBenefits'] = df['TotalPayBenefits'].fillna(0)
```

```python
df.isnull()
```

# Fourth Request

```python
plt.figure(figsize=(10,6))
plt.hist(df['TotalPay'], bins=20, color='blue')
plt.title('TotalPay Distribution')
plt.xlabel('TotalPay')
plt.ylabel('Number of Employees')
plt.show()

# Pie chart for departments
plt.figure(figsize=(10,6))
df['Year'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Employees in Different Years')
plt.show()
```

# Fifth Request

```python
print('mean total pay for each job title:', '\n', df.groupby('JobTitle')['TotalPay'].mean())
print('mean total pay for each year:','\n', df.groupby('Year')['TotalPay'].mean())
```

# Sixth Request

```python
df['Benefits'].corr(df['TotalPayBenefits'])
```

```python
plt.figure()
plt.scatter(df['Benefits'], df['TotalPayBenefits'])
plt.title('Scatter plot of Benefits with TotalPayBenefits')
plt.ylabel('TotalPayBenefits')
plt.xlabel('Benefits')
plt.show()
```
